```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("data/data_kotor.csv")
df.head()
```

```python
print(df.isnull().sum())
```

```python
missing_rows = df[df.isnull().any(axis=1)]
print(missing_rows.head())
```

```python
df_clean = df.copy()
df_clean['Peak'] = df['Peak'].astype(str).str.replace(r'\[.*?\]','',regex=True)
print(df_clean["Peak"])
```

```python
df_clean['Peak'] = df_clean['Peak'].astype(float)
df_clean['Peak'].fillna(df_clean['Peak'].mean())
print(df_clean["Peak"])
```

# Deteksi data ganda

```python
print(df_clean.duplicated().sum(), "duplicate rows found.")

# Show the actual duplicated records
duplicates = df_clean[df_clean.duplicated()]
print(duplicates)

df_clean = df_clean.drop_duplicates()
print(df_clean)
```
